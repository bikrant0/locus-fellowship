## 1. Creating Arrays

```python
import numpy as np

# Create an array from a list
scores = np.array([85, 90, 78, 92, 60])
print("Array:", scores)
```

```python
# Basic info about an array
print("Shape:", scores.shape)      # number of elements
print("Data type:", scores.dtype)  # type of values stored
print("Dimensions:", scores.ndim)  # 1 for a simple list-like array
```

```python
# Common way: array of zeros
zeros = np.zeros(5)
print("zeros:", zeros)
```

```python
# Common way: array of ones
ones = np.ones(5)
print("ones:", ones)
```

```python
# Common way: a range of values
sequence = np.arange(0, 10, 2)   # values from 0 to 10, step 2
print("range:", sequence)
```

## 2. Indexing and Slicing

Indexing starts at 0. Slicing uses [start:end], where end is not included.

```python
scores = np.array([85, 90, 78, 92, 60])

# Indexing - get one value by its position
print("First element:", scores[0])
print("Last element:", scores[-1])
```

```python
# Slicing - get a range of values
print("First 3 elements:", scores[0:3])
print("Elements from index 2 onward:", scores[2:])
```

```python
# Boolean filtering - get only the values that match a condition
print("Values above 80:", scores[scores > 80])
```

## 3. Math Operations

Operations apply to every element at once. No loops are needed.

```python
scores = np.array([85, 90, 78, 92, 60])

# Applying an operation to every element at once
print("Add 5 to all:", scores + 5)
print("Multiply all by 2:", scores * 2)
```

```python
# Two arrays of the same size can be combined element by element
bonus = np.array([2, 1, 3, 0, 5])
print("Add two arrays together:", scores + bonus)
```

## 4. Aggregate Functions

Functions that summarize an entire array into a single value.

```python
scores = np.array([85, 90, 78, 92, 60])

print("Sum:", np.sum(scores))
print("Mean:", np.mean(scores))
print("Max:", np.max(scores))
print("Min:", np.min(scores))
```

## 5. Reshape

Reshape changes the shape of an array without changing its data.

```python
numbers = np.arange(1, 13)   # 1 to 12
print("Original:", numbers)
print("Reshaped to 3x4:\n", numbers.reshape(3, 4))
```
